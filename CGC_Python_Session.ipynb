```
def pattern_1(n):
    
    for i in range(1,n+1):
        print("*"*i)
        
    print("")

    #Pattern 2
    for i in range(1,n+1):
        print(" "*(n-i),end='') #Spaces
        print("*"*(2*i-1)) #Stars
    
pattern_1(5)
```
